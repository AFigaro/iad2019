##### Визуализация:

Немного коснемся визуализации построим простые графики. В начале заполним пропуски в поле price

```python
df1['price'].fillna(0, inplace=True)
```

Теперь посмотрим на распределение цен в зависимости от класса:

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot') #зададим стиль
```

```python
df_grouped = data.groupby('Pclass')['Fare'].mean()
plt.figure(figsize=(10, 6))
plt.bar(df_grouped.index, df_grouped.values, tick_label=['Class 1', 'Class 2', 'Class 3'])
plt.xticks(rotation=45) #делаем подписи по оси x вертикальными
plt.title('Dependency between price and class of the cabin') #запишем название графика
plt.xlabel('Pclass') #назовем ось x
plt.ylabel('Mean price') #назовем ось y
plt.show()
```

Теперь посмотрим на зависимость между ценой на билет и возрастом пассажира. Также в качестве цвета пассажира выведем индикатор, который показывает выжил человек или нет:

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Fare'],c=data['Survived'], label=data['Survived'])
plt.title('Dependency between age and fare of the ticket') #запишем название графика
plt.xlabel('Age') #назовем ось x
plt.ylabel('Fare') #назовем ось y
plt.colorbar()
plt.show()
```
